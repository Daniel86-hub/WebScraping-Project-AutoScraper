# Web Scraping 

```python
from autoscraper import AutoScraper
```

```python
amazon_url="https://www.amazon.com/s?k=iphone"
```

```python
wanted_list=['$198.00','Apple iPhone 8, 64GB, Space Gray - Fully Unlocked (Renewed)']
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_fxoe':'Title','rule_fbku':'Price'})
scraper.keep_rules(['rule_fxoe','rule_fbku'])
scraper.save('amazon-search')
```

```python
import pandas as pd
result=scraper.get_result_similar("https://www.amazon.com/s?k=iphone",group_by_alias=True)
Item=pd.DataFrame(result['Title'])
Item
```

```python
Price=pd.DataFrame(result['Price'])
Price
```

```python
Iphone_Quote=pd.concat([Item[0], Price[0]], axis=1, keys=['Phone_Model', 'Price'])
Iphone_Quote
```

```python
Iphone_Quote.dropna(axis=0)
```
